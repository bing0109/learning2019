```python
values =['nan','00000','000000','NO CLUE']
data=pd.read_csv('data/311-service-requests.csv',dtype={'Incident Zip':np.str},na_values={'Incident Zip':values})
data.head()
```

```python
data['Incident Zip'].unique()
```

```python
data['Incident Zip']=data['Incident Zip'].str.slice(0,5)
data['Incident Zip'].unique()
```

美国或者加拿大的邮编，都是以0开头、以1开头，或者是以3开头

```python
is_0=data['Incident Zip'].str.startswith('0',na=False)
is_1=data['Incident Zip'].str.startswith('1',na=False)
is_3=data['Incident Zip'].str.startswith('3',na=False)
data.[~(is_0 | is_1 |is_3)]=np.nan
data['Incident Zip'].unique()
```
