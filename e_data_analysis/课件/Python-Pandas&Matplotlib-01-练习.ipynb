**作业：**

1.将'311-service-requests.csv'数据集加载到DataFrame中，查看数据集；

2.取出'Complaint Type'对应的列；

3.将'Complaint Type'列中的数值频数计算，并进行排序；(提示：求频数，用value_counts()函数)

4.针对排序好的数据，取出Top5；

5.针对Top5画出柱状图，并将图形保存。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('/home/zelin/311-service-requests.csv',dtype={'Incident Zip':str})
c_counts = data['Complaint Type'].str.title().value_counts()
# type(c_counts)
c_counts.sort_values(ascending=False)
c_top5=c_counts.head()
# len(data['Complaint Type'].str.title().unique())
```

```python
# 用pyplot画图
fig = plt.figure(figsize=(9,6))
ax = fig.subplots()
c_bar = ax.bar(c_top5.index,c_top5,color=['r','g','y','purple','gray'])
plt.title('Complaint Type',fontsize=18)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xticks(rotation=45,fontsize=12)
plt.yticks(fontsize=12)

for x,y in zip(c_top5.index,list(c_top5)):
#     plt.text(x,y+150,str('%.2f'%(y/1000))+'k',fontsize=12,horizontalalignment='center')
#     plt.text(x,y+150,'%.2f'%y,fontsize=12,horizontalalignment='center')
    plt.text(x,y+150,y,fontsize=12,horizontalalignment='center')
    
# str(b),fontsize=10,horizontalalignment='center'
```
