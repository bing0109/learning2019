# 概率论

# 1. 概率的基本概念

在概率论里面，对于随机现象，我们把可以产生不确定，但是可以明确结果的过程定义为**随机试验**，常常可以成为**试验**。
* 可以重复发生；
* 不止一种结果，但需要知道所有可能产生的结果；
* 产生的结果不确定

随机试验有很多种，例如常出现的掷骰子，摸球，射击，抛硬币等。所有的随机试验的结果可以分为两类来表示：
* 数量化表示：射击命中的次数，商场每个小时的客流量，每天经过某个收费站的车辆等，这些结果本身就是数字；
* 非数量化表示：抛硬币的结果（正面/反面），化验的结果（阳性/阴性）等，这些结果是定性的，非数量化的。但是可以用示性函数来表示，例如可以规定正面（阳性）为1，反面（阴性）为0，这样就实现了非数量化结果的数量化表示。

## 1.1 样本空间和事件

### 1.1.1 样本空间

**样本空间：**
试验所有可能结果的组成的一个集合。

抛掷一枚硬币的样本空间：S = {正面，反面}

检测一个零件的样本空间：S = {合格，不合格}

抛掷一枚色子的样本空间：S = {1,2,3,4,5,6}

### 1.1.2 计数法则

**多步骤试验的计数法则：**如果一个试验可以分成循环的k个步骤，第一步中有$n_1$种试验结果，第二步中有$n_2$种试验结果…以此类推。那么所有可能的试验结果的总数为$n_1$×$n_2$×…×$n_k$。

可以通过**树型图**来帮助分析多步骤试验的样本空间。例如考虑一个抛掷两枚硬币的试验。

**组合计数法则：**从N项中任取n项的组合数为$$C_{N}^{n}=\frac{N!}{n!(N-n)!}$$

**排列计数法则：**从N项中任取n项的排列数为$$P_{N}^{n}=\frac{N!}{(N-n)!}$$

### 1.1.3 概率分配

**概率分配的基本条件**
1. 分配给的每个实验结果的概率值必须在0到1之间；若每个试验的结果表示为$E_i$，$P(E_i)$表示i种结果发生的概率，则有：$$0≤P(E_i)≤1，(对所有i)$$
2. 所有试验结果之和必须为1.对于有n种试验结果的情形，有：$$P(E_1)+P(E_2)+…+P(E_n)=1$$

常用的概率分配方法有**古典法**、**几何法**、**相对频数**和**主观概率法**。

**古典法：**当各种试验结果是等概率发生时，适合采用古典法进行概率分配。例如掷色子，每种结果的概率都为1/6。

**相对频数法：**适用于试验可以大量重复，并且取得试验结果发生比率的数据。

**主观法：**不能确定试验结果等可能，或者无法取得相关数据，可以考虑采用此方法。

**几何法：**碰到试验结果在某个区间的可能性时，可以考虑用几何法。

> 思考：假如一对情侣约定在9:00~10:00之间碰面，如果等待超过20分钟，就不等了。求情侣之间碰面的可能性？

## 1.2 事件及其概率

**事件**是样本点的一个集合。事件是样本空间的子集，同样是一个集合；

**事件的概率**等于事件中所有样本点的概率之和。

## 1.3 概率的基本性质

### 1.3.1 事件的补

给定一个事件A，定义**事件A的补**为所有不包含在事件A里面的样本点，记为$A^c$。可以通过**文氏图**来说明补的概念。

在任何概率应该用中,事件A和事件A的$A^c$补必有一个发生：$$P(A)+P(A^c)=1$$

因此可以**利用事件A的补计算概率**：$$P(A)=1-P(A^c)$$

### 1.3.2 加法公式

**两个事件的并：**A和B的并是所有的属于A或B，或同时属于二者的样本点构成的事件，记做A∪B。

**两个事件的交：**A和B的并是所有的属于A和B的样本点构成的事件，记做A∩B。

**加法公式：**$$P(A∪B) = P(A) + P(B) - P(A∩B)$$

> 思考：一个有50人的小厂，每个工人必须按时完成定额，并且装配的产品需要通过最终检验。在期末绩效测评的时候，生产经理发现50名工人中有5人没能按时完成工作，6人装配的产品不合格，2人既没有按时完成工作又存在装配不合格问题。出现一项问题的概率是多少？

**互斥事件**：两个事件之间没有公共样本点。

**互斥事件的加法公式：**$$P(A∪B) = P(A) + P(B)$$

## 1.4 条件概率

设A, B是两个事件，且P(B)>0， 则在事件B发生的条件下，事件A发生的条件概率(conditional probability)为：$$P(A | B)=\frac{P(A∩B)}{P(B)}$$

在条件概率中，最本质的变化是样本空间缩小了——由原来的整个样本空间缩小到了给定条件的样本空间。

**乘法公式**是由条件概率公式所得，可以获得两个事件的并。$$P(A∩B) = P(B)·P(A|B) = P(A)·P(B|A)$$

对于多个事件的并，可以由**乘法公式的推广**。对于任何正整数n≥2，当$P(A_1∩A_2∩…∩A_{n-1}) > 0$时，有：$$P(A_1∩A_2∩...∩A_{n-1}) = P(A1)P(A2|A1)P(A3|A1∩A2)…P(An|A1∩A2∩…∩An-1)$$

## 1.5 全概率公式

### 1.5.1 全概率

**前提假设：**设B1，B2，....为有限或无限个事件，它们两两互斥且在每次试验中至少发生一个，即：
* 不重，Bi ∩ Bj = ∅（不可能事件）i≠j ,
* 不漏，B1∪B2∪.... = S（必然事件）。

这时，称事件组 B1, B2,...是样本空间S的一个划分，把具有这些性质的一组事件称为一个“**完备事件组**”。如下图所示：

**全概率公式（formula of total probability):**设 $B_1$, $B_2$,...，$B_n$是样本空间S的一个划分，A为任一事件(上图中红色圈内)，则：$$P(A) = \sum_{i=1}^{n}P(B_i)P(A | B_i)$$

也可以分为两步来看全概率公式：

设$P(B_j) = p_j$, $P(A|B_j) = q_j$, j = 1, 2, ..., n，则$$P(A) = \sum_{i=1}^{n}p_j*q_j$$

在运用全概率公式时的已知未知条件为：
* 划分后的每个小事件的概率，即P(Bi), i = 1, 2, ..., n；
* 每个小事件发生的条件下，A发生的概率，即P(A|Bi), i = 1, 2, ..., n；
* 求解目标是计算A发生的概率，即P(A)。

### 1.5.2 意义

全概率公式的意义在于，当直接计算P(A)较为困难，而$P(B_i)，P(A|B_i)$ (i=1,2,...)的计算较为简单时，可以利用全概率公式计算P(A)。思想就是，将事件A分解成若干个小事件，通过求每个小事件的概率，然后相加从而求得事件A的概率。

而将事件A进行分割的时候，不是直接对A进行分割，而是先找到样本空间S的一个划分$B_1,B_2,...,B_n$,这样事件A就被事件$A∩B_1,A∩B_2,...A∩B_n$分解成了n部分，即$A = A∩B_1 + A∩B_2 + ... + A∩B_n$, 每一Bi发生都可能导致A发生相应的概率是$P(A|B_i)$，由加法公式得$$P(A) = P(A∩B_1) + P(A∩B_2) + ... + P(A∩B_n)= P(B1)P(A|B_1) + P(B_2)P(A|B_2) + ... +P(B_n) P(A|B_n)$$

## 1.6 贝叶斯定理

与全概率公式解决的问题相反，贝叶斯公式是建立在条件概率的基础上寻找事件发生的原因（即大事件A已经发生的条件下，分割中的小事件$B_i$在A发生的条件下的概率），设$B_1,B_2,...,B_n$是样本空间S的一个划分，则对任一事件A（P(A)>0），有$$P(B_i|A) = \frac{P(B_i)P(A|B_i)}{\sum_{j=1}^{n}P(B_j)P(A|B_j)}$$

上式即为**贝叶斯公式(Bayes formula)**
$B_i$ 常被视为导致试验结果A发生的“原因”，$P(B_i)(i=1,2,...,n)$表示各种原因发生的可能性大小，故称先验概率（权重）；$P(B_i|A)(i=1,2,...,n)$则反映当试验产生了结果A之后，再对各种原因概率的新认识，故称后验概率。

> **问题1：**设某公路经过的货车与客车的数量之比为1:2，货车中途停车修车的概率为0.02，客车为0.01，今有一辆汽车中途停车修理，则该车是货车的概率是多少？

> **问题2：** 装有10件某产品（乒乓球）（其中一等品5件，二等品3件，三等品2件）的箱子中丢失一件产品，但不知是几等品，今从箱中任取2件产品，结果都是一等品，则丢失的也是一等品的概率为？

# 2. 随机变量

在几乎所有的教材中，介绍概率论时都是从事件和样本空间说起的，但是后面的概率论都是围绕着随机变量展开的。可以说前面的事件和样本空间都是引子，引出了随机变量这个概率论中的核心概念。后面的统计学是建立在概率论的理论基础之上的，因此可以说理解随机变量这个概念是学习和运用概率论与数理统计的关键。

## 2.1 定义

简单来说，随机变量就是对一个试验结果的数值描述。

设随机试验的样本空间是S。若对S中的每个样本点e，都有唯一的实数值X(e)与之对应，则称X(e)为随机变量，简记为X。

## 2.2 随机变量与事件

随机变量的本质是一种函数（映射关系），在古典概率模型中，“事件和事件的概率”是核心概念；但是在现代概率论中，“随机变量及其取值规律”是核心概念。

随机变量实际上只是事件的另一种表达方式，这种表达方式更加形式化和符号化，也更加便于理解以及进行逻辑运算。不同的事件，其实就是随机变量不同取值的组合。

>对于随机试验，我们所关心的往往是与所研究的特定问题有关的某个或某些量，而这些量就是随机变量。当然，有时我们所关心的是某个或某些特定的随机事件。例如，在特定一群人中，年收入在万元以上的高收入者，以及年收入在3000元以下的低收入者，各自的比率如何？这看上去像是两个孤立的事件。可是，若我们引入一个随机变量X：$$X=随机抽出一个人其年收入$$
则X是我们关心的随机变量。上述两个事件可分别表示为{X>10000}或{X<3000}。这就看出：随机事件这个概念实际上包容在随机变量这个更广的概念之内。也可以说，随机事件是从静态的观点来研究随机现象，而随机变量则是一种动态的观点，一如数学分析中的常量与变量的区分那样，变量概念是高等数学有别于初等数学的基础概念。同样，概率论能从计算一些孤立事件的概率发展为一个更高的理论体系，其基本概念就是随机变量。

总结：
* 随机变量与之前定义的事件是有关系的，因为每个样本点本身就是一个基本事件；
* 无论是数量化的结果还是非数量化的结果，即不管试验结果是否与数值有关, 都可以引入变量, 使试验结果与数建立对应关系；
* 随机变量本质上是一种函数，其目的就是建立试验结果（样本空间中的点，同基本事件e）与实数之间的对应关系（例如将"正面"映射为1，"反面"映射为0）；
* 自变量为基本事件e，定义域为样本空间S，值域为某个实数集合，多个自变量可以对应同一个函数值，但不允许一个自变量对应多个函数值；
* 随机变量X取某个值或某些值就表示某种事件，且具有一定的概率；
* 随机变量中的随机来源于随机试验结果的不确定性

## 2.3 随机变量的分类

随机变量从其可能取的值全体的性质可以分为两大类：离散型随机变量和连续型随机变量。

更加深入的理解随机变量的分类点击[这里](https://en.wikipedia.org/wiki/Probability_distribution)。

### 2.3.1 离散型随机变量

可以取有限多个值或者无限多个值的随机变量成为**离散型随机变量**。
离散型随机变量的取值在整个实数轴上是间隔的，要么只有有限个取值，要么是无限可数的。

### 2.3.2 连续性随机变量

可以取某一区间或多个区间内任意值的随机变量称为**连续性随机变量**。

连续型随机变量的取值要么包括整个实数集(−∞,+∞)，要么在一个区间内连续，总之这类随机变量的可能取值要比离散型随机变量的取值多得多，它们的个数是无限不可数的。

# 3 概率分布

随机变量的**概率分布**是描述随机变量取不同值的概率。

## 3.1 离散型概率分布

对于离散型随机变量x,**概率函数**给出随机变量取每种值的概率，记做f(x)。

**离散型概率的基本条件：**$$f(x)≥0$$
$$\sum f(x)=1$$

**离散型随机变量的期望：** $$E(x) = μ= \sum xf(x)$$

**离散型随机变量的方差：** $$Var(x) = σ^2 = \sum (x-μ)^2f(x)$$

### 3.1.1 伯努利分布

**定义：**伯努利分布应该是所有分布里面最简单的分布，也是二项分布的基本单元。其样本空间中只有两个点，一般取为{0,1}。不同的伯努利分布只是取到这两个值的概率不同。如果将抛一次硬币看作是一次伯努利实验，且将正面朝上记做1，反面朝上记做0。那么伯努利分布中的参数p就表示硬币正面朝上的概率，其期望同参数p，也是关注硬币正面朝上的概率。

若伯努利试验成功，则伯努利随机变量取值为1。若伯努利试验失败，则伯努利随机变量取值为0。记其成功概率为p(0≤p≤1)，失败概率为q=1−p。则其概率质量函数（PMF）为：$$f(x) = x*p + (1-x)*(1-p)$$   其中，x=0或1

伯努利分布只有一个参数p，记做X ∼ Bernoulli(p)X，或X ∼ B(1,p)，读作X服从参数为p的伯努利分布

### 3.1.2 二项分布

如果把一个伯努利分布独立的重复n次，就得到了一个二项分布。二项分布是最重要的离散型概率分布之一。随机变量X要满足这个分布有两个重要条件：
* 各次试验的条件是稳定的；
* 各次试验之间是相互独立的。

还是利用抛硬币的例子来比较伯努利分布和二项分布：如果将抛一次硬币看作是一次伯努利实验，且将正面朝上记做1，反面朝上记做0。那么抛n次硬币，记录正面朝上的次数Y，Y就服从二项分布。假如硬币是均匀的，Y的取值应该大部分都集中在n/2附近，而非常大或非常小的值都很少。由此可见，二项分布关注的是计数，伯努利分布关注的是比值（正面朝上的计数/n）。

二项分布有两个参数——试验次数n和每次试验成功的概率p。**二项概率质量函数为**：$$f(x) = (_{x}^{n})p^x(1-p)^(n-x) = \frac{n!}{x!(n-x)!}p^x(1-p)^{(n-x)}$$其中，x为试验成功的次数，n为试验的次数，p为一次试验成功的概率，f(x)为n次试验有x次成功的概率。

一个随机变量X服从参数为n和p的二项分布，记做X∼Binomial(n,p)，或X∼B(n,p)

**二项分布的数学期望和方差：**$$E(x) = μ =np$$
$$Var(x) = σ^2 = np(1-p)$$

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']

def binom_pmf(n=1, p=0.1):
    binom_dis = stats.binom(n, p)
    x = np.arange(binom_dis.ppf(0.0001), binom_dis.ppf(0.9999))
    print(x)  # [ 0.  1.  2.  3.  4.]
    fig, ax = plt.subplots(1, 1)
    ax.plot(x, binom_dis.pmf(x), 'bo', label='伯努利的PMD')
    ax.vlines(x, 0, binom_dis.pmf(x), colors='b', lw=5, alpha=0.5)
    ax.legend(loc='best', frameon=False)
    plt.ylabel('概率')
    plt.title('伯努利概率分布(n={}, p={})'.format(n, p))
    #plt.show()
    
binom_pmf(n=100, p=0.9)
```

### 3.1.3 泊松分布

日常生活中，大量事件的发生是有固定频率的。例如某医院平均每小时出生3个婴儿，某网站平均每分钟有2次访问等。它们的特点就是，我们可以预估这些事件在某个时间段内发生的总次数，但是没法知道具体的发生时间。已知平均每小时出生3个婴儿，请问下一个小时，会出生几个？有可能一下子出生6个，也有可能一个都不出生。这是我们没法知道的。

如果某事件以固定频率λ，随机且独立地出现，该事件在单位时间内出现的次数（个数）可能看出是服从泊松分布。泊松分布是一个常用的离散型随机分布，主要估计在特定的时间段或空间中，某件事发生的次数。

泊松分布有一个参数λ（有的地方表示为μ），表示单位时间（或单位面积）内随机事件的平均发生次数，**泊松分布PMF**表示为：$$f(x)=\frac{λ^x e^{-λ}}{x!}$$其中，λ为事件在一个区间内发生次数的期望，f(x)为事件在某一区间发生x次的概率。

一个随机变量X服从参数为λ的泊松分布，记做X∼Poisson(λ)，或X∼P(λ)。

**泊松分布的数学期望和方差：**泊松分布的一个重要性质就是数学期望和方差相等，都等于λ。

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']

def poisson_pmf(mu=3):
    poisson_dis = stats.poisson(mu)
    x = np.arange(poisson_dis.ppf(0.00001), poisson_dis.ppf(0.999999))
    fig, ax = plt.subplots(1, 1)
    ax.plot(x, poisson_dis.pmf(x), 'bo', ms=8, label='poisson pmf')
    ax.vlines(x, 0, poisson_dis.pmf(x), colors='b', lw=5, alpha=0.5)
    ax.legend(loc='best', frameon=False)
    plt.ylabel('概率')
    plt.title('泊松分布的PMF(mu={})'.format(mu))
    plt.show()

poisson_pmf(mu=3)
```

### 3.1.4 超几何分布

对于超几何分布，用符号N表示总体容量，r表示总体容量中成功的个数，采用不放回抽样，从总体中选取n个元素，计算x个成功的概率。**超几何的概率质量函数：**$$f(x) = \frac{(_x^r)(_{n-x}^{N-r})}{(_n^N)}$$

**超几何分布的数学期望和方差：**$$E(x) = μ = n(\frac{r}{N})$$
$$Var(x) = σ^2 = n(\frac{r}{N})(1-\frac{r}{N})(\frac{N-n}{N-1})$$

当总体容量足够大的时候，超几何分布近似二项分布。至于代码实现可以参照Scipy的[官方文档](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.hypergeom.html#scipy.stats.hypergeom)。

## 3.2 连续型概率分布

对于连续性随机分布函数，不能像离散型变量一样给出x值直接计算概率，连续型变量通过给定区间上曲线f(x)下面的面积该出在该区间的概率，这条曲线f(x)就叫做**概率密度函数**。

由于在任一特定点上曲线f(x)下面积为零，所以连续型变量概率定义意味着它取某一点的概率为零。

**连续型概率的基本条件：**$$f(x)≥0$$
$$\int _{-∞}^{+∞} f(x)dx=1$$

**连续性概率的数学期望：**$$E(x) = μ = \int xf(x)dx$$

**连续性概率的方差：**$$Var(x) = σ^2 = \int (x-μ)^2f(x)dx$$

### 3.2.1 均匀分布

均匀分布算是最简单的连续型概率分布。因为其概率密度是一个常数，不随随机变量X取值的变化而变化。

均匀分布具有等可能性，也就是说，服从 U(a,b)U(a,b) 上的均匀分布的随机变量 XX 落入 (a,b)(a,b) 中的任意子区间上的概率只与其区间长度有关，与区间所处的位置无关。

**均匀概率密度函数：**$$f(x)=\begin{cases} 
		\dfrac {1}{b-a},&a≤x≤b\\	
		0,&其它		
	\end{cases}$$

如果连续型随机变量 X 的具有均匀概率密度函数，则称 X 服从 [a,b] 上的均匀分布（uniform distribution）,记作 X∼U(a,b) 或 X∼Unif(a,b)

**均匀分布的数学期望和方差：**$$E(x) = \frac{a+b}{2}$$
$$Var(x) = \frac{(b-a)^2}{12}$$

从定义可以看出来，定义一个均匀分布需要两个参数，定义域区间的起点 a 和终点 b，但是在Python中是 location 和 scale, 分别表示起点和区间长度。

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']
def uniform_distribution(loc=0, scale=1):
    uniform_dis = stats.uniform(loc=loc, scale=scale)
    x = np.linspace(uniform_dis.ppf(0.01),uniform_dis.ppf(0.99), 100)
    fig, ax = plt.subplots(1, 1)
    # 直接传入参数
    ax.plot(x, stats.uniform.pdf(x, loc=2, scale=4), 'r-',
            lw=5, alpha=0.6, label='均匀分布的PDF')
    # 从冻结的均匀分布取值
    ax.plot(x, uniform_dis.pdf(x), 'k-',
            lw=2, label='冻结的PDF')

    # 计算ppf分别等于0.001, 0.5, 0.999时的x值
    vals = uniform_dis.ppf([0.001, 0.5, 0.999])
    print(vals)  # [ 2.004  4.     5.996]

    # Check accuracy of cdf and ppf
    print(np.allclose([0.001, 0.5, 0.999], uniform_dis.cdf(vals)))  # Ture

    r = uniform_dis.rvs(size=10000)
    ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
    plt.ylabel('概率')
    plt.title(r'均匀分布概率密度函数({}, {})'.format(loc, loc+scale))
    ax.legend(loc='best', frameon=False)
    plt.show()

uniform_distribution(loc=2, scale=4)
```

### 3.2.2 指数分布

其实指数分布和离散型的泊松分布之间有很大的关系。泊松分布表示单位时间（或单位面积）内随机事件的平均发生次数，指数分布则可以用来表示独立随机事件发生的时间间隔。

指数概率分布可以用于描述诸如装载一辆卡车需要的时间、高速公路上两起事故之间的距离等随机变量。

**指数概率密度函数：**$$f(x) = λe^{-λx},\ x≥0$$其中λ为期望。

记做 X∼E(λ)或 X∼Exp(λ)。

**指数分布的数学期望和方差：**$$E(x) = \frac{1}{λ}$$
$$Var(x) = \frac{1}{λ^2}$$

指数分布的一个显著的特点是其具有无记忆性。例如如果排队的顾客接受服务的时间长短服从指数分布，那么无论你已经排了多久时间的队，在排 t 分钟的概率始终是相同的。用公式表示就是：$$P(X≥s+t|X≥s)=P(X≥t)$$

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']
def exponential_dis(loc=0, scale=1.0):
    """
    指数分布，exponential continuous random variable
    按照定义，指数分布只有一个参数lambda，这里的scale = 1/lambda
    :param loc: 定义域的左端点，相当于将整体分布沿x轴平移loc
    :param scale: lambda的倒数，loc + scale表示该分布的均值，scale^2表示该分布的方差
    :return:
    """
    exp_dis = stats.expon(loc=loc, scale=scale)
    x = np.linspace(exp_dis.ppf(0.000001),
                    exp_dis.ppf(0.999999), 100)
    fig, ax = plt.subplots(1, 1)

    # 直接传入参数
    ax.plot(x, stats.expon.pdf(x, loc=loc, scale=scale), 'r-',
            lw=5, alpha=0.6, label='指数的PDF')

    # 计算ppf分别等于0.001, 0.5, 0.999时的x值
    vals = exp_dis.ppf([0.001, 0.5, 0.999])
    print(vals)  # [ 2.004  4.     5.996]

    # Check accuracy of cdf and ppf
    print(np.allclose([0.001, 0.5, 0.999], exp_dis.cdf(vals)))

    r = exp_dis.rvs(size=10000)
    ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
    plt.ylabel('概率')
    plt.title('指数分布的概率密度函数({})'.format(round(1/scale,2)))
    ax.legend(loc='best', frameon=False)
    plt.show()

exponential_dis(loc=0, scale=15)
```

### 3.2.4 正态分布

正态分布也许是出现频率最高的分布，其他人对正态分布的熟悉程度应该也是所有分布中最高的。由于中心极限定理的存在，正态分布也是所有分布应用最广泛的分布，没有之一。

密度若随机变量 X 的概率密度符合下面的形式，就称 X 服从参数为 μ,σ的正态分布（或高斯分布），记为  $X∼N(μ,σ^2)$。**正态分布的概率密度函数：**$$f(x)=\dfrac{1}{\sqrt{2π}σ} e^{(x-μ)^2/2σ^2}$$

其中，μ为均值，σμ为标准差。

如果上面公式中 μ=0,σ=1，就叫做标准正态分布，一般记做 Z∼N(0,1)。由于标准正态分布在统计学中的重要地位，它的累积分布函数（CDF）有一个专门的表示符号：Φ。

**正态分布有四条性质：**
* f(x) 关于 x=μ 对称；
* 当x≤μ时，f(x)是严格单调递增函数；
* $f_{max}=f(μ)=\dfrac{1}{\sqrt{2π}σ}$；
* 当 $X∼N(μ,σ^2)$ 时，$\dfrac{X−μ}{σ}∼N(0,1)$。在计算一般正态分布的概率时，可以转化为标准正态分布函数来计算。

下面来看看不同参数下的正太分布：

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False
def diff_normal_dis():
    """
    不同参数下的指数分布
    """
    norm_dis_0 = stats.norm(0, 1)  # 标准正态分布
    norm_dis_1 = stats.norm(0, 0.5)
    norm_dis_2 = stats.norm(0, 2)
    norm_dis_3 = stats.norm(2, 2)

    x0 = np.linspace(norm_dis_0.ppf(1e-8), norm_dis_0.ppf(0.99999999), 1000)
    x1 = np.linspace(norm_dis_1.ppf(1e-10), norm_dis_1.ppf(0.9999999999), 1000)
    x2 = np.linspace(norm_dis_2.ppf(1e-6), norm_dis_2.ppf(0.999999), 1000)
    x3 = np.linspace(norm_dis_3.ppf(1e-6), norm_dis_3.ppf(0.999999), 1000)
    fig, ax = plt.subplots(1, 1)
    ax.plot(x0, norm_dis_0.pdf(x0), 'r-', lw=2, label=r'miu=0, sigma=1')
    ax.plot(x1, norm_dis_1.pdf(x1), 'b-', lw=2, label=r'miu=0, sigma=0.5')
    ax.plot(x2, norm_dis_2.pdf(x2), 'g-', lw=2, label=r'miu=0, sigma=2')
    ax.plot(x3, norm_dis_3.pdf(x3), 'y-', lw=2, label=r'miu=2, sigma=2')
    plt.ylabel('概率')
    plt.title('不同参数下的正态分布')
    ax.legend(loc='best', frameon=False)
    plt.show()

diff_normal_dis()
```

由上图可知：
* 当固定 σ，改变 μ 的大小时，f(x) 图形的形状不变，只是沿着 x 轴作平移变换，因此 μ 被称为位置参数（决定对称轴的位置）；
* 当固定 μ，改变 σ 的大小时， f(x) 图形的对称轴不变，形状改变， σ 越小，图形越高越瘦， σ 越大，图形越矮越胖，因此 σ 被称为尺度参数（决定曲线的分散程度）

### 3.2.5 二项概率的正太近似

当实验次数很大时，求解二项概率函数比较困难，可以根据中心极限定理（后面会对该定理进行说明），近似的等价正态分布。在np≥5和n(1-p)≥5情况下，近似正态分布，正态曲线中取μ=np和$σ=\sqrt{np(1-p)}$。

下面通过函数实现比较一下：

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import math

plt.rcParams['font.sans-serif'] = ['KaiTi']
def binom2normal_dis(n=100,p=0.1):
    binom_dis = stats.binom(n,p)
    x1 = np.arange(binom_dis.ppf(0.0001), binom_dis.ppf(0.9999))
    norm_dis = stats.norm(n*p,math.sqrt(n*(1-p)*p))
    x2 = np.linspace(norm_dis.ppf(0.00000001), norm_dis.ppf(0.99999999), 1000)
    
    fig, ax = plt.subplots(1, 1)
    ax.plot(x1, binom_dis.pmf(x1), 'bo', label='二项分布B({},{})'.format(n,p))
    ax.vlines(x1, 0, binom_dis.pmf(x1), colors='b', lw=5, alpha=0.5)
    ax.plot(x2, norm_dis.pdf(x2), 'r-', lw=2, label='正太分布N({},{})'.format(n*p,math.sqrt(n*(1-p)*p)))
    ax.legend(loc='best', frameon=False)
    plt.ylabel('概率')
    plt.title('二项概率正太近似')

n=100
p=0.1
binom2normal_dis(n,p)
print(stats.binom(n,p).pmf(10))
print(stats.norm(n*p,math.sqrt(n*(1-p)*p)).pdf(10))
```

# 4. 大数定理和中心极限定理

大数定理和中心极限定理这两个定理可以说是概率论中最重要的两个定理。也是由于中心极限定理的存在，使得正态分布从其他众多分布中脱颖而出，成为应用最为广泛的分布。这两个定理在概率论的历史上非常重要，因此对于它们的研究也横跨了几个世纪（始于18世纪初），众多耳熟能详的大数学家都对这两个定理有自己的贡献。因此，这两个定理都不是单一的定理。不同的大数定理和中心极限定理从不同的方面对相同的问题进行了阐述，它们条件各不相同，得到的结论的强弱程度也不一样。

## 4.1 大数定理(Law of Large Numbers，LLN)

设$X_1,X_2,…,X_n$是独立同分布的随机变量，记它们的公共均值为μ。又设它们的方差存在并记为$σ^2$。则对任意给定的ε>0，有$$\lim_{n \to +\infty}P(|\overset{-}X_n −μ|≥ε)=0$$

这个式子指出了“当n很大时，$\overset{-}X_n$接近μ”的确切含义。这里的“接近”是概率上的，也就是说虽然概率非常小，但还是有一定的概率出现意外情况（例如上面的式子中大于ε的情况）。只是这样的可能性越来越小，这样的收敛性，在概率论中叫做“$\overset{-}X_n$依概率收敛于μ”。

下面用程序模拟抛硬币的过程来辅助说明大数定律：

```python
import random
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']
def flip_plot(minExp, maxExp):
    # 两个参数的含义，抛硬币的次数为2的minExp次方到2的maxExp次方，也就是一共做了(2**maxExp - 2**minExp)批次实验，每批次重复抛硬币2**n次

    ratios = []
    xAxis = []
    for exp in range(minExp, maxExp + 1):
        xAxis.append(2**exp)
    for numFlips in xAxis:
        numHeads = 0 # 初始化，硬币正面朝上的计数为0
        for n in range(numFlips):
            if random.random() < 0.5:  # random.random()从[0, 1)随机的取出一个数
                numHeads += 1  # 当随机取出的数小于0.5时，正面朝上的计数加1
        numTails = numFlips - numHeads  # 得到本次试验中反面朝上的次数
        ratios.append(numHeads/float(numTails))  #正反面计数的比值
    plt.title('正面/反面比率')
    plt.xlabel('试验批次')
    plt.ylabel('正面/反面')
    plt.plot(xAxis, ratios)
    plt.hlines(1, 0, xAxis[-1], linestyles='dashed', colors='r')
    plt.show()

flip_plot(4, 16)
```

用random模块生成区间\[0,1)之间的随机数，如果生成的数小于0.5，就记为硬币正面朝上，否则记为硬币反面朝上。由于random.random()生成的数可以看做是服从区间[0,1)上的均匀分布，所以以0.5为界限，随机生成的数中大于0.5或小于0.5的概率应该是相同的(相当于硬币是均匀的)。这样就用随机数模拟出了实际的抛硬币试验。理论上试验次数越多(即抛硬币的次数越多)，正反面出现的次数之比越接近于1(也就是说正反面各占一半).

**两个重要的不等式：**

**马尔科夫不等式：**
若Y为只取非负值得随机变量，则对任何常数$\varepsilon>0$有$$P(Y≥ε)≤E(Y)/ε$$

**切比雪夫不等式：**若Var(Y)存在，则$$P(|Y-E(Y)|≥ε)≤Var(Y)/ε^2$$

## 4.2 中心极限定理（Central Limit Theorem，CLT）

从总体抽取容量为n的样本，当样本容量很大的时候，样本均值的抽样分布，近似服从正太分布。

设$X_1,X_2,\cdots,X_n$为独立同分布的随机变量，$E(X_i)=μ,Var(X_i)=σ^2,0<σ^2<∞$。则对任何实数x，有$$\lim_{n \to \infty}P\left( \dfrac{1}{\sqrt{n}σ} (X_1+\cdots+X_n-nμ)≤ε\right)=Φ(x)$$

这里Φ(x)是标准正太分布N(0,1)的分布函数。

在一般情况下，我们很难求出$X_1+X_2+\cdots+X_n$的分布的确切形式，但当n很大时，可以通过Φ(x)给出其近似值。

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

plt.rcParams['font.sans-serif'] = ['KaiTi']

def sampling2pmf(n, dist, t=10000):
    """
    n: sample size for each experiment
    t: how many times do you do experiment, fix in 10000
    dist: frozen distribution
    """
    ber_dist = dist
    sum_of_samples = []
    for i in range(t):
        samples = ber_dist.rvs(size=n)  # 与每次取一个值，取n次效果相同
        sum_of_samples.append(np.sum(samples))
    val, cnt = np.unique(sum_of_samples, return_counts=True)
    pmf = cnt / len(sum_of_samples)
    return val, pmf


def plot(n, dist, subplot, plt_handle):
    """
    :param n: sample size
    :param dist: distribution of each single sample
    :param subplot: location of sub-graph, such as 221, 222, 223, 224
    :param plt_handle: plt object
    :return: plt object
    """
    bins = 10000
    plt = plt_handle
    plt.subplot(subplot)
    mu = n * dist.mean()
    sigma = np.sqrt(n * dist.var())
    samples = sampling2pmf(n=n, dist=dist)
    plt.vlines(samples[0], 0, samples[1],
               colors='g', linestyles='-', lw=3)
    plt.ylabel('Probability')
    plt.title('Sum of bernoulli dist. (n={})'.format(n))
    # normal distribution
    norm_dis = stats.norm(mu, sigma)
    norm_x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, bins)
    pdf1 = norm_dis.pdf(norm_x)
    plt.plot(norm_x, pdf1, 'r--')
    return plt

size = [1, 4, 20, 80, 200, 1000]

# sum of bernoulli distribution
dist_type = 'bern'
bern_para = [0.4]
single_sample_dist = stats.bernoulli(p=bern_para[0])  # 定义一个伯努利分布

# 下面是利用matplotlib画图
plt.figure(1)
plt = plot(n=size[0], dist=single_sample_dist, subplot=321, plt_handle=plt)
plt = plot(n=size[1], dist=single_sample_dist, subplot=322, plt_handle=plt)
plt = plot(n=size[2], dist=single_sample_dist, subplot=323, plt_handle=plt)
plt = plot(n=size[3], dist=single_sample_dist, subplot=324, plt_handle=plt)
plt = plot(n=size[4], dist=single_sample_dist, subplot=325, plt_handle=plt)
plt = plot(n=size[5], dist=single_sample_dist, subplot=326, plt_handle=plt)
plt.tight_layout()
plt.savefig('sum_of_{}_dist.png'.format(dist_type), dpi=200)
plt.show()
```

由上图可以看到，增加试验次数和在x轴上分割区间的个数后，随机变量之和与正态分布之间的重合度随着样本量的增加而升高。这也很容易解释：对于随机变量和的分布，抽样次数（试验次数）越多，最终画出来的图越能代表整个分布。但是抽样次数本身却不影响该分布的类型和参数，就像在学校研究全体同学的身高，抽样的人数不会影响身高的真实分布，只会影响我们利用所得的样本描绘出来的分布的形状。
